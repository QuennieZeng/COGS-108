### Looking at obesity rates

We will focus on obesity rates and see if there is one state that has a higher obesity rate compared to another.

```python
# create new dataframe that only has obesity rates and state
obesity_df = state_merged_df[['Obesity', 'state']]
obesity_df 
```

After looking at this, we can compare the obesity rates using a barplot.

```python
import matplotlib.pyplot as plt

# Sort the DataFrame by obesity rates
obesity_df = obesity_df.sort_values(by='Obesity')

# Get the highest and lowest obesity rates
highest_obesity = obesity_df['Obesity'].max()
lowest_obesity = obesity_df['Obesity'].min()

# Plotting
plt.figure(figsize=(10, 10))

# Plot bars for all states
plt.barh(obesity_df['state'], obesity_df['Obesity'], color='skyblue')

# Highlight the bar for the highest obesity rate
plt.barh(obesity_df.loc[obesity_df['Obesity'] == highest_obesity, 'state'], highest_obesity, color='red')

# Highlight the bar for the lowest obesity rate
plt.barh(obesity_df.loc[obesity_df['Obesity'] == lowest_obesity, 'state'], lowest_obesity, color='green')

plt.xlabel('Obesity Rate (%)')
plt.title('Obesity Rates by State')
plt.show()

# Find the state with the lowest obesity rate
state_lowest_obesity = obesity_df.loc[obesity_df['Obesity'] == lowest_obesity, 'state'].values[0]
print("Highest obesity: ", highest_obesity, ", state: ", obesity_df.loc[obesity_df['Obesity'] == highest_obesity, 'state'].values[0])
print("Lowest obesity: ", lowest_obesity, ", state: ", state_lowest_obesity)
```

We can see visually that the state with the highest obesity rate is the top red bar, Louisiana, and the state with the lowest obesity rate is the bottom green bar, Colorado. The range of the obesity rates in America per state is from around 20% to greater that 35%. There are visibly no outliers but we can also check this through code. 

```python
# Calculating for outliers
# Calculate the interquartile range (IQR)
Q1 = obesity_df['Obesity'].quantile(0.25)
Q3 = obesity_df['Obesity'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify potential outliers
outliers = obesity_df[(obesity_df['Obesity'] < lower_bound) | (obesity_df['Obesity'] > upper_bound)]

print("Potential outliers:")
print(outliers)
```

We see here that the data has no outliers. We thus learn that the obesity rates per state are at a range of around 15% with Colorado having the lowest rate and Louisiana having the highest rate.
```python
# Imports
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import statsmodels.api as sm

from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
import pandas as pd
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt

# Assuming state_df is your original DataFrame
state_df_copy = state_merged_df.copy()

# Set the size of the figure
plt.figure(figsize=(14, 10))  # Width = 14 inches, Height = 10 inches

# Create a scatterplot
plt.scatter(state_df_copy['All Fast Food Restaurants'], state_df_copy['Obesity'])

# Calculate the linear regression line
slope, intercept, r_value, p_value, std_err = linregress(state_df_copy['All Fast Food Restaurants'], state_df_copy['Obesity'])
x_values = np.array([min(state_df_copy['All Fast Food Restaurants']), max(state_df_copy['All Fast Food Restaurants'])])
y_values = slope * x_values + intercept

# Plot the regression line
plt.plot(x_values, y_values, color='red', label='Regression Line')

# Add state names as text annotations 
for state, x, y in zip(state_df_copy['state'], state_df_copy['All Fast Food Restaurants'], state_df_copy['Obesity']):
    plt.text(x + 0.1, y + 0.1, state)

# Add labels and title
plt.xlabel('Fast Food Restaurants per 100k People')
plt.ylabel('Obesity (% of Population)')
plt.title('Fast Food Restaurants per 100k People vs. Obesity Rate by State')

# Add legend
plt.legend()

# Show plot
plt.show()

```
